```python
from lxml import etree
import random
import requests
```

This model takes two words as input and returns a sentence of length n composed of predicted words that follow the inputed two words. The algorithm analyses a text and creates a mapping dictionary that stores the predictions that could come after two words. First, a function that creates this dictionary and stores two word tuples as keys and a list of unique predicted words as values is required.

```python
def construct_dictionary(text):
  """
  This function takes some text as input and constructs a dictionary of two word tuples as keys and lists of possible next words as values.
  """
  words = text.split()
  mapping_dict = {}

  for i in range(len(words)-2):
    if (words[i], words[i+1]) in mapping_dict.keys():
      if words[i+2] not in mapping_dict[(words[i], words[i+1])]:
        mapping_dict[(words[i], words[i+1])].append(words[i+2])
    else:
      mapping_dict[(words[i], words[i+1])] = [words[i+2]]

  return mapping_dict
```

```python
"""
Unit test
"""
assert construct_dictionary("earthquakes are natural disasters") == {('earthquakes', 'are'): ['natural'], ('are', 'natural'): ['disasters']}
```

Next is a function that randomly returns a word from a list, this will help to construct a sentence based on random words.

```python
def predict_next_word(possible_words):
  """
  This function takes a list of words as input and returns a random word from the list.

  There is no way to create a unit test for this function as it is generating a random integer.
  """
  index = random.randint(0,len(possible_words)-1)
  return possible_words[index]
```

Next, I implemented a way to get text data from a Wikipedia page using the help of Chat GPT. This function will return the content of the page to help create the mapping dictionary for future use.

```python
def get_wikipedia_text(title):
    """
    Prompted Chat GPT to give a function that scrapes a Wikipedia page and returns the content of the page as a string.
    """
    url = "https://en.wikipedia.org/w/api.php"

    headers = {
        "User-Agent": "MyWikipediaApp/1.0 (your_email@example.com)"
    }

    params = {
        "action": "query",
        "prop": "extracts",
        "explaintext": True,
        "titles": title,
        "format": "json"
    }

    response = requests.get(url, headers=headers, params=params)
    response.raise_for_status()
    data = response.json()

    pages = data["query"]["pages"]
    page = next(iter(pages.values()))

    return page.get("extract", "")
```

Next is a function that is needed to construct the sentence. This function joins the predicted words with the sentence that is being created until it reaches a desired length.

```python
def construct_sentence(num_of_words, sentence, mapping_dict):
  """
  This function takes a list of words and number of words and a mapping dictionary of two words with their possible predictions as input and returns
  an extended sentence as a string created by predicting and appending words based on the last two words of the sentence up until the specified
  number of words or until the last two words don't have any other possible predictions.

  Promted Chat GPT using what is written in the doc string alongside the previous code to gain an understanding of how the code needs to be structured.

  A unit test is not possible for this function as it calls a function that uses the random module.
  """
  for i in range(num_of_words):
    key = (sentence[i], sentence[i+1])
    if key not in mapping_dict:
        break
    list_of_possible_words = mapping_dict[key]
    next_word = predict_next_word(list_of_possible_words)
    sentence.append(next_word)



  return " ".join(sentence)
```

Finally, some text is obtained from a Wikipedia page concerning earthquakes and the user is asked to input two words of their choice and how many words they want to predict. A sentence containing the two words and the correct number of following words will be shown as long as both words are keys in the mapping dictionary. If not, two random words will be chosen for the user, followed by the number of predicted words they inputed.

```python
text = get_wikipedia_text("Earthquake")

mapping_dict = construct_dictionary(text)

first_word = input('Please enter the first word: ')
second_word = input('Please enter the second word: ')
num_of_words = int(input('Please enter the how many words do you want to predict: '))

if (first_word, second_word) in mapping_dict.keys():
  print(construct_sentence(num_of_words, [first_word, second_word], mapping_dict))

else:
  print("Words not found. I will pick two random start words.")
  words = list(mapping_dict.keys())
  two_words = words[random.randint(0,3)]
  first_word = two_words[0]
  second_word = two_words[1]

  print(construct_sentence(num_of_words, [first_word, second_word], mapping_dict))


```
